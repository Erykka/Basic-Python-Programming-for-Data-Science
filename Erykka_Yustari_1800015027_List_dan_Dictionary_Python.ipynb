2. Buatlah sebuah dictionary dengan jumlah elemen 100 dan lakukanlah beberapa modifikasi(minimal 3) di dalamnya.

```python
#Tuliskan kodemu di bawah ini:

#Membuat Dictionary
a=100
b=dict()
for x in range (1,a+1):
  b[x]=x*x
print(b)
print(len(b))
```

```python
#Modifikasi 1
a=100
v=dict()
for x in range (1,a+1):
  v[x]=2*x*(x+1)
print(v)
print(len(v))
```

```python
#Modifikasi 2
a=100
l=dict()
for x in range (1,a+1):
  l[x]=(x+1)*x
print(l)
print(len(l))
```

```python
#Modifikasi 3
a=100
l=dict()
for x in range (1,a+1):
  l[x]=(x+11)*(x+12)
print(l)
print(len(l))
```
